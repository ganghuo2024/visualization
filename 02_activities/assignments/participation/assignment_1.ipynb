```python
 import numpy as np
 import matplotlib.pyplot as plt
 import pandas as pd
 import scipy
 import PIL
 import requests

 np.random.seed(613)
 x1 = np.arange(50)
 y1 = np.random.randint(0, 75,50)
 x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
 y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Create the figure with two subplots in one row
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))

# Scatter plot in the first subplot
ax1.scatter(x1, y1, marker='o', color='red')
ax1.set_title('Scatter Plot')

# Bar plot in the second subplot
ax2.bar(x2, y2, color='yellow')
ax2.set_title('Bar Plot')

plt.tight_layout()
plt.show()
```

```python
# Arrange subplots using plt.subplot_mosaic()
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
 ['ax2', 'ax3']],
 figsize=(7, 4))
for label, ax in someaxes.items():
 ax.text(0.5, 0.5, f'This is {label!r}',
 fontsize=14,
 ha = 'center',
 transform=ax.transAxes)
```

```python
# Arrange subplots using plt.subplot_mosaic()
# Add large x axis titles to the subplots using contrained layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4),
                                    layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel("A big label!", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize= 18)
someaxes["ax3"].set_xlabel("third label", fontsize = 18)
```

```python
# Plot multiple viz on one axes object

# Make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# Define our figure and axes
fig, ax = plt.subplots(figsize=(7, 3))

# Call both bar and line elements to the same ax
ax.bar(x, y1, color='indigo', label='Y1')
ax.plot(x, y2, color='red', label='Y2', marker='o')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
# Add error bars in our plot

# Calculate the standard deviation of our data
y2_sd = np.std(y2)

# Plot the line chart
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = 'red')

# Add error bars to the line chart
ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick = 4)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# Define our data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# Getting the image from the internet
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)

# Create the figure and plot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y, color="red", label="Line Plot")
ax.legend()

# Overlay an image axis on the figure
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])  # Adjust these values as needed
ax_image.imshow(image)
ax_image.axis('off')  # Turn off axis for the image

plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests 
```

```python
# Load "tips" dataset
tips = sns.load_dataset("tips")
print (tips)
```

```python
# Make a line plot using seaborn using premade style
sns.set_style('white')

sns.lineplot(data = tips,
             x='total_bill',
             y= 'tip')
```

```python
# Add title and axis label to the line plot
tipgraph = sns.lineplot(data = tips,
                        x='total_bill',
                        y='tip')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip amount ($)')
```

```python
# Modify aesthetic elements of the line plot
fig, ax = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(
    data=tips,
    x='total_bill',
    y='tip',
    color='hotpink',
    linestyle='--',
    linewidth=3,
    marker='o',  # Use a valid marker
    markerfacecolor='indigo',
    ax=ax
)
```

```python
# Create a scatter plot using multiple variables
tipgraph = sns.scatterplot(
    data=tips, 
    x='total_bill',
    y='tip', 
    style='time', 
    hue='day',
    palette=['purple', 'yellowgreen', 'orange', 'blue']  # Include enough colors for each unique 'day'
)

tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'  # Corrected the typo
)
```

```python
# Create pairplot to explore the relationships between variables
sns.pairplot(data = tips, 
             hue = 'day')
```

```python
# Use replot to explore tip vs total_bill by day and sex
daysplot = sns.relplot(data=tips,
                       x="total_bill",
                       y="tip",
                       hue="sex",
                       col="day",
                       kind="scatter",
                       col_wrap=2)
```

```python
# Import the plotly package and create sample data
import plotly.graph_objects as go 

x1 = np.array(["luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
# Make our figure
graph = go.Figure()

# Choose our plot type
graph.add_trace(go.Bar(x=x1, y=y1))
```

```python
# Update our layout to include titles
graph.update_layout(title="Pirate Scores", 
                    xaxis_title="Pirates",
                    yaxis_title="Score")
graph.show()
```

```python
# Customizing plotly graphs
graph = go.Figure()

graph.add_trace(go.Scatter(
    x=x1, 
    y=y1, 
    mode='markers',
    marker=dict(
        size=10,  # Numeric value for size
        color='hotpink',  # Color of the markers
        opacity=1,
        line=dict(
            width=2,  # Numeric value for width
            color='purple'
        )
    )
))

graph.update_layout(
    title='Interactive Pirate Plot',
    xaxis_title='Pirates',
    yaxis_title='Scores',
    width=500, 
    height=500
)

graph.show()
```

```python
# Import the wordcloud package andload the sample dataset of movie quotes
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines='skip')

# Combine all quotes into a single string
text = " ".join(quote for quote in df['quote'])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Combine all quotes into a single string
text = "".join(each for each in df.quote)

# Generate the word cloud
wordcloud = WordCloud(background_color='white', colormap='inferno').generate(text)

# Create a plot
fig, ax = plt.subplots(figsize=(7, 3))

# Display the word cloud
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")  # Turn off the axis

plt.show()
```
